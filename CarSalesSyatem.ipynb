```python
%%writefile cars.cpp
#include <iostream>
#include <vector>
#include <string>
using namespace std;

// ===== CLASS YA GARI =====
class Car {
private:
    string make;
    string model;
    int year;
    double price;

public:
    Car(string mk, string mdl, int yr, double pr) { // contructor
        make = mk;
        model = mdl;
        year = yr;
        price = pr;
    }

    void display() const {
        cout << "Make: " << make << ", Model: " << model
             << ", Year: " << year << ", Price: $" << price << endl;
    }

    string getMake() const { return make; }
    string getModel() const { return model; }
    int getYear() const { return year; }
    double getPrice() const { return price; }
};

// ===== CLASS YA INVENTORY =====
class Inventory {
private:
    vector<Car> cars;

public:
    void addCar(const Car& car) {
        cars.push_back(car);
    }

    void listCars() const {
        if (cars.empty()) {
            cout << "No cars in inventory.\n";
            return;
        }

        cout << "\n--- Cars in Inventory ---\n";
        for (const auto& car : cars) {
            car.display();
        }
    }

    void searchByMake(const string& make) const {
        bool found = false;
        for (const auto& car : cars) {
            if (car.getMake() == make) {
                car.display();
                found = true;
            }
        }
        if (!found) cout << "No cars found for make: " << make << endl;
    }
};

// ===== MAIN FUNCTION =====
int main() {
    Inventory inventory;

    int choice;
    do {
        cout << "\n====== Car Sales Management System ======\n";
        cout << "1. Add Car\n";
        cout << "2. View Cars\n";
        cout << "3. Search by Make\n";
        cout << "0. Exit\n";
        cout << "Enter your choice: ";
        cin >> choice;

        if (choice == 1) {
            string make, model;
            int year;
            double price;

            cout << "Enter car make: ";
            cin >> make;
            cout << "Enter car model: ";
            cin >> model;
            cout << "Enter year: ";
            cin >> year;
            cout << "Enter price: ";
            cin >> price;

            Car newCar(make, model, year, price);
            inventory.addCar(newCar);
            cout << "Car added successfully!\n";
        }
        else if (choice == 2) {
            inventory.listCars();
        }
        else if (choice == 3) {
            string make;
            cout << "Enter make to search: ";
            cin >> make;
            inventory.searchByMake(make);
        }
        else if (choice == 0) {
            cout << "Exiting system...\n";
        }
        else {
            cout << "Invalid choice. Try again.\n";
        }

    } while (choice != 0);

    return 0;
}
```
